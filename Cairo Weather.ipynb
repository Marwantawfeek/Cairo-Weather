```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('Cairo Weather2021.csv')
```

```python
df
```

Time	Temperature (° F)			Dew Point (° F)			Humidity (%)			Wind Speed (mph)			Pressure (Hg)			Precipitation (in)


```python
# Converting Temp from F to C
df['MaxTemp']=5/9 * (df['MaxTemp']-32)
df['AvgTemp']=5/9 * (df['AvgTemp']-32)
df['MinTemp']=5/9 * (df['MinTemp']-32)
```

```python
# Converting Dew Point from F to C
df['MaxDewPoint']=5/9 * (df['MaxDewPoint']-32)
df['AvgDewPoint']=5/9 * (df['AvgDewPoint']-32)
df['MinDewpoint']=5/9 * (df['MinDewpoint']-32)
```

```python
# Converting wind speed from mph to kph
df['MaxWindSpeed']=1.609344*df['MaxWindSpeed']
df['AvgWindSpeed']=1.609344*df['AvgWindSpeed']
df['MinWindSpeed']=1.609344*df['MinWindSpeed']
```

```python
# Dropping empty Column
df.drop(['Precipitation (in)'], axis=1,inplace=True)
```

```python
# Converting date datatype from object to Date
df['Date'] =  pd.to_datetime(df['Date'])
```
